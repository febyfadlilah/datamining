# Tugas 2 - Diskritisasi

Diskritisasi adalah proses merubah fitur-fitur kontinyu tersebut menjadi fitur-fitur kategorikal

```python
import pandas as pd
import math
```

```python
ds = pd.read_csv("https://raw.githubusercontent.com/febyfadlilah/dataset/main/IRIS.csv")
ds.describe()
```

Langkah yang pertama adalah dengan mencari Equal Width Intervals yaitu membagi rentang dengan menggunakan nilai min dan max.

$w=\frac{Zmax-Zmin}{k}$

dimana
 
k = jumlah pengelompokkan

**Sepal_Length**

```python
# mengubah diskritisasi sepal_length
def iris(min, max):
  k = 4
  w = (max - min)/k
  return (w)
```

```python
iris(4.3, 7.9)
```

Untuk melakukan diskritisasi maka diperlukan rentang. Untuk menghitung rentang dapat dicari dengan menggunakan

$v = nilai min +w.i$

dimana nilai min merupakan batas minimal, w adalah rentang dan i merupakan increment

```python
#rumus rentang
def rentang_sepal_length(i):
  min = 4.3
  v = min + i * iris(4.3,7.9)
  return (v)
```

```python
rentang_sepal_length(1)
```

```python
rentang_sepal_length(2)
```

```python
rentang_sepal_length(3)
```

```python
rentang_sepal_length(4)
```

```python
for col in ["sepal_length"]: 
  nilai = ds[col]
  def equal(nilai):
    if nilai >= 4.3 and nilai < 5.2:
      return "A"
    elif nilai >= 5.2 and nilai < 6.1:
      return "B"
    elif nilai >= 6.1 and nilai < 7.0:
      return "C"
    elif nilai >= 7.0 and nilai < 7.9:
      return "D"
```

```python
ds["sepal_length"] = ds["sepal_length"].apply(equal)
ds[["sepal_length"]]
```

**Sepal_Width**

```python
iris(2,4.4)
```

```python
#rumus rentang
def rentang_sepal_width(i):
  min = 2
  v = min + i * iris(2,4.4)
  return (v)
```

```python
rentang_sepal_width(1)
```

```python
rentang_sepal_width(2)
```

```python
rentang_sepal_width(3)
```

```python
rentang_sepal_width(4)
```

```python
for col1 in ["sepal_width"]: 
  nilai1 = ds[col1]
  def equal_sepal_width(nilai1):
    if nilai1 >= 2 and nilai1 < rentang_sepal_width(1) :
      return "A"
    elif nilai1 >= rentang_sepal_width(1) and nilai1 < rentang_sepal_width(2):
      return "B"
    elif nilai1 >= rentang_sepal_width(2) and nilai1 < rentang_sepal_width(3):
      return "C"
    elif nilai1 >= rentang_sepal_width(3) and nilai1 < rentang_sepal_width(4):
      return "D"
```

```python
ds["sepal_width"] = nilai1.apply(equal_sepal_width)
ds[["sepal_width"]]
```
