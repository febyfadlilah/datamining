#UAS - Prediksi Jenis Obat

```python
import pandas as pd
```

Read Data

```python
df = pd.read_csv("https://raw.githubusercontent.com/febyfadlilah/dataset/main/drug200.csv")
df
```

Exploration Data

```python
df.head()
```

```python
df[["Age","Na_to_K"]].agg(['min','max'])
# df[["precipitation", "temp_max", "temp_min", "wind"]].agg(['min','max'])
```

```python
df.Drug.value_counts()
```

Preprocessing Data

Normalize Data

```python
data2 = df[["Sex"]]
data2
```

```python
jk = pd.get_dummies(data2)
jk
```

```python
data3 = df[["BP"]]
data3
```

```python
tekanan = pd.get_dummies(data3)
tekanan
```

```python
data4 = df[["Cholesterol"]]
data4
```

```python
kolestrol = pd.get_dummies(data4)
kolestrol
```

```python
dataOlah = pd.concat([jk,tekanan,kolestrol], axis=1)
```

```python
dataHasil = pd.concat([df,dataOlah], axis = 1)
dataHasil
```

```python
X = dataHasil.drop(columns=["Drug","Sex","BP","Cholesterol"])
y = dataHasil.Drug
```

```python
X
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y)
y = le.transform(y)
y
```

```python
le.inverse_transform(y)
```

```python
labels = pd.get_dummies(dataHasil.Drug).columns.values.tolist()
labels
```

Normalisasi MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
X
```

```python
X.shape, y.shape
```

Split Data

```python
# membagi data menjadi data testing(20%) dan training(80%)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from numpy import array
metode1 = KNeighborsClassifier(n_neighbors=3)
metode1.fit(X_train, y_train)
print(metode1.score(X_train, y_train))
print(metode1.score(X_test, y_test))
y_pred = metode1.predict(scaler.transform(array([[0.13,0.51,1,0,0,1,0,1,0]])))
le.inverse_transform(y_pred)[0]
```

Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

metode2 = GaussianNB()
metode2.fit(X_train, y_train)
print(metode2.score(X_train, y_train))
print(metode2.score(X_test, y_test))
y_pred = metode2.predict(array([[0.13,0.51,1,0,0,1,0,1,0]]))
le.inverse_transform(y_pred)[0]
```

Decission Tree

```python
from sklearn import tree

metode3 = tree.DecisionTreeClassifier(criterion="gini")
metode3.fit(X_train, y_train)
print(metode3.score(X_train, y_train))
print(metode3.score(X_test, y_test))
y_pred = metode3.predict(array([[0.13,0.51,1,0,0,1,0,1,0]]))
le.inverse_transform(y_pred)[0]
```

Export 



*   Label Encoder

```python
from sklearn.utils.validation import joblib
# label encoder
joblib.dump(le, "le.save") 

# scaler
joblib.dump(scaler, "scaler.save") 

# model
joblib.dump(metode1, "nb.joblib")
joblib.dump(metode2, "knn.joblib")
joblib.dump(metode3, "tree.joblib")
```
