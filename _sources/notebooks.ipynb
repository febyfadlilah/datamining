**Menghitung Biner Simetris**

Untuk menghitung jarak biner simetris maka diperlukan rumus

---
$$Rumus: d(i,j) = \ \frac{r+s} {q+r+s+t} $$

---

$$Keterangan: $$

* r: i=1, j=0
* s: i=0, j=1 
* q: i=1, j=1
* t: i=0, j=0







```python
def binersimetris(i,j):
  q=r=s=t=0
  for col in ['gender','test_preparation_course']:
    if data[col][i]==1 and data[col][j] == 1:
      q+=1
    elif data[col][i]==1 and data[col][j]==0:
      r+=1
    elif data[col][i]==0 and data[col][j] == 1:
      s+=1
    else :
      t+=1
  return (r+s)/(q+r+s+t)
```

### **Menghitung Data Nominal**

Atribut nominal mempunyai k kategori yang tidak memiliki urutan yang berarti. Jarak antara dua objek dengan m variabel adalah
$$d(i,j) = \frac{m-p}{p}$$
Dengan p adalah banyaknya variabel yang mempunyai atribut yang sama.
$$Keterangan : $$

* m = jumlah data yang dibandingkan
* P = jumlah data sama

```python
data[["race","parental_level"]].head()
```

```python
# Menghitung jarak data Nominal
def d1(i,j):
    P = 4
    M = 0
    for col in ["race","parental_level"]:
        if data[col][i] == data[col][j]:
            M += 1

    return (P-M)/P
```

```python
# Menampilkan hasil perhitungan
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

Dalam perhitungan kali ini menggunakan rumus Euclidian Distance untuk menemukan jarak pada type data numeric dengan menggunakan
$$d(i,j)=\sqrt{(x1-xj1)^2 + (xi2 - xj2)^2+.....}$$

```python
# Menghitung jarak Numeric dengan metode Euclidean
def d2(x,y,h):
    result = 0
    for col in [hasil1, hasil2, hasil3]:
        result = math.pow(abs(col[x] - col[y]), h)
    return result**(1/h)
```

```python
# Hasil perhitungan
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```

Langkah yang pertama adalah dengan mencari Equal Width Intervals yaitu membagi rentang dengan menggunakan nilai min dan max.
$$w=\frac{Zmax-Zmin}{k}$
dimana 
k = jumlah pengelompokkan

**Sepal_Length**

```python
# mengubah diskritisasi sepal_length
def iris(min, max):
  k = 4
  w = (max - min)/k
  return (w)
```

```python
iris(4.3, 7.9)
```

Untuk melakukan diskritisasi maka diperlukan rentang. Untuk menghitung rentang dapat dicari dengan menggunakan 
$Rumus:$$v = nilai min +w.i$
dimana nilai min merupakan batas minimal, w adalah rentang dan i merupakan increment

```python
#rumus rentang
def rentang_sepal_length(i):
  min = 4.3
  v = min + i * iris(4.3,7.9)
  return (v)
```

# *Naive Bayes Classifier*

Naive Bayes adalah metode yang cocok untuk klasifikasi biner dan multiclass. Metode yang juga dikenal sebagai Naive Bayes Classifier ini menerapkan teknik supervised klasifikasi objek di masa depan dengan menetapkan label kelas ke instance/catatan menggunakan probabilitas bersyarat.


**Naive Bayes**

---

$Rumus:$
$$P(xⅈ|𝑦ⅈ)=\frac{1}{\sqrt2πδ^2}^e-\frac{(xi.μj)^2}{2δ^2j}$$

---


$$δ=\sqrt\frac{Σ(xi-μ)^2}{N}$$

---

$$μ=\frac{Σᴺᵢ=1}{N}$$

---
$$P(x|c)P(c)$$
$Example:$
1.   P(x | Setosa) = cari semua standar deviasinya dan kalikan semua standar deviasinya
2.   P(Setosa) = Jumlah setosa dibagi jumlah data ($\frac{50}{100}$)
3. P= P(x | setosa) * P(Setosa)




```python
# Naive Bayes Classification

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/febyfadlilah/dataset/main/IRIS.csv")
```

```python
#looking at the first 5 values of the dataset
dataset.head()
```

```python
#Spliting the dataset in independent and dependent variables
X = dataset.iloc[:,:4].values
y = dataset['species'].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 82)
```

```python
# Feature Scaling to bring the variable in a single scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting Naive Bayes Classification to the Training set with linear kernel
from sklearn.naive_bayes import GaussianNB
nvclassifier = GaussianNB()
nvclassifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = nvclassifier.predict(X_test)
print(y_pred)
```

```python
#lets see the actual and predicted value side by side
y_compare = np.vstack((y_test,y_pred)).T
#actual value on the left side and predicted value on the right hand side
#printing the top 5 values
y_compare[:5,:]
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#finding accuracy from the confusion matrix.
a = cm.shape
corrPred = 0
falsePred = 0

for row in range(a[0]):
    for c in range(a[1]):
        if row == c:
            corrPred +=cm[row,c]
        else:
            falsePred += cm[row,c]
print('Correct predictions: ', corrPred)
print('False predictions', falsePred)
print ('\n\nAccuracy of the Naive Bayes Clasification is: ', corrPred/(cm.sum()))    
```

## *K-Means Clustering*

K-means merupakan salah satu algoritma yang bersifat unsupervised learning. K-Means memiliki fungsi untuk mengelompokkan data kedalam data cluster. Algoritma ini dapat menerima data tanpa ada label kategori.

K means clustering merupakan metode algoritma dasar,yang diterapkan sebagai berikut

* Menentukan jumlah cluster
* Secara acak mendistribusikan data cluster
* Menghitung rata rata dari data yang ada di cluster.
* Menggunakan langkah baris 3 kembali sesuai nilai treshold
* Menghitung jarak antara data dan nilai centroid(K means clustering)
*Distance space dapat diimplementasikan untuk menghitung jarak data dan centroid. Contoh penghitungan jarak yang sering digunakan adalah manhattan/city blok distance



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


#visualisasi 
import seaborn as sns
import matplotlib.pyplot as pyplot
# sns.set(Style="darkgrid")

import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
from sklearn.cluster import KMeans
wcss =[]
for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
  kmeans.fit(df_iris)
  wcss.append(kmeans.inertia_)

#plotting graph
plt.plot(range(1,11), wcss)
plt.title ('The albow method')
plt.xlabel('Number of Cluster')
plt.ylabel('WCSS')
plt.show()
```

```python
#Applying
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(df_iris)
```
