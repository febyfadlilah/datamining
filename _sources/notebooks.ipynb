### **Menghitung Biner**



```python
data[["gender","test_preparation_course"]].head()
```

```python
#code gendernya
cowo = "male"
cewe = "female"
lulus = "completed"
tdk_lulus = "none"

#nilai konversi biner
value_of_one = 1
value_of_zero = 0

def change_biner(gender):
  return value_of_one if gender == cowo else value_of_zero
def change_binertes(test_preparation_course):
  return value_of_one if test_preparation_course == lulus else value_of_zero
```

```python
#Update Nilai dari Gender dan Test Preparation
data['gender'] = data['gender'].apply(change_biner)
data['test_preparation_course']= data['test_preparation_course'].apply(change_binertes)
```

```python
data[["gender","test_preparation_course"]].head()
```

**Menghitung Biner Simetris**

```python
def binersimetris(i,j):
  q=r=s=t=0
  for col in ['gender','test_preparation_course']:
    if data[col][i]==1 and data[col][j] == 1:
      q+=1
    elif data[col][i]==1 and data[col][j]==0:
      r+=1
    elif data[col][i]==0 and data[col][j] == 1:
      s+=1
    else :
      t+=1
  return (r+s)/(q+r+s+t)
```

```python
binersimetris(1,2)
```

```python
binersimetris(1,3)
```

```python
binersimetris(1,4)
```

**Menghitung Biner Asimetris**

```python
def binerii(i,j):
  q=r=s=0
  for col in ['gender','test_preparation_course']:
    if data[col][i]==1 and data[col][j] == 1:
      q+=1
    elif data[col][i]==1 and data[col][j]==0:
      r+=1
    elif data[col][i]==0 and data[col][j] == 1:
      s+=1
  return (r+s)/(q+r+s)
```

### **Menghitung Data Nominal**

```python
data[["race","parental_level"]].head()
```

```python
# Menghitung jarak data Nominal
def d1(i,j):
    P = 4
    M = 0
    for col in ["race","parental_level"]:
        if data[col][i] == data[col][j]:
            M += 1

    return (P-M)/P
```

```python
# Menampilkan hasil perhitungan
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

### **Menghitung Data Numeric**

```python
data[["math_score","reading_score","writing_score"]].head()
```

```python
data.describe()
```

```python
#normalisasi math score
std1 = 15.16308
mean1 = 66.08900
for col1 in ['math_score']:
  data1 = data[col1]
  hasil1 = (data1-mean1)/std1
  print(hasil1)
#normalisasi reading_score
std2 = 14.600192
mean2 = 68.054000
for col2 in ['reading_score']:
  data2 = data[col2]
  hasil2 = (data2-mean2)/std2
  print(hasil2)

#normalisasi writing_score
std3 = 15.195657
mean3 = 68.054000
for col3 in ['writing_score']:
  data3 = data[col3]
  hasil3 = (data3-mean3)/std3
  print(hasil3)
```

```python
# Menghitung jarak Numeric dengan metode Euclidean
def d2(x,y,h):
    result = 0
    for col in [hasil1, hasil2, hasil3]:
        result = math.pow(abs(col[x] - col[y]), h)
    return result**(1/h)
```

```python
# Hasil perhitungan
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```

**Sepal_Length**

```python
# mengubah diskritisasi sepal_length
def iris(min, max):
  k = 4
  w = (max - min)/k
  return (w)
```

```python
iris(4.3, 7.9)
```

```python
#rumus rentang
def rentang_sepal_length(i):
  min = 4.3
  v = min + i * iris(4.3,7.9)
  return (v)
```

```python
rentang_sepal_length(1)
```

```python
rentang_sepal_length(2)
```

```python
rentang_sepal_length(3)
```

```python
rentang_sepal_length(4)
```

```python
for col in ["sepal_length"]: 
  nilai = ds[col]
  def equal(nilai):
    if nilai >= 4.3 and nilai < 5.2:
      return "A"
    elif nilai >= 5.2 and nilai < 6.1:
      return "B"
    elif nilai >= 6.1 and nilai < 7.0:
      return "C"
    elif nilai >= 7.0 and nilai < 7.9:
      return "D"
```

```python
ds["sepal_length"] = ds["sepal_length"].apply(equal)
ds[["sepal_length"]]
```

**Sepal_Width**

```python
iris(2,4.4)
```

```python
#rumus rentang
def rentang_sepal_width(i):
  min = 2
  v = min + i * iris(2,4.4)
  return (v)
```

```python
rentang_sepal_width(1)
```

```python
rentang_sepal_width(2)
```

```python
rentang_sepal_width(3)
```

```python
rentang_sepal_width(4)
```

```python
for col1 in ["sepal_width"]: 
  nilai1 = ds[col1]
  def equal_sepal_width(nilai1):
    if nilai1 >= 2 and nilai1 < rentang_sepal_width(1) :
      return "A"
    elif nilai1 >= rentang_sepal_width(1) and nilai1 < rentang_sepal_width(2):
      return "B"
    elif nilai1 >= rentang_sepal_width(2) and nilai1 < rentang_sepal_width(3):
      return "C"
    elif nilai1 >= rentang_sepal_width(3) and nilai1 < rentang_sepal_width(4):
      return "D"
```

```python
ds["sepal_width"] = nilai1.apply(equal_sepal_width)
ds[["sepal_width"]]
```
